```python
def parse_ctc(val):
    if pd.isna(val):
        return np.nan

    # Convert to string and clean spaces/suffixes
    s = str(val).upper().replace("LPA", "").replace(" ", "").strip()

    # Remove commas to extract raw numeric structure safely
    s_clean = s.replace(",", "")

    try:
        num = float(s_clean)
        # Trap 2 Check: If it's written out fully as raw INR currency (e.g., 1550000)
        # convert it down to Lakhs Per Annum (LPA) by dividing by 100,000
        if num >= 100000:
            return num / 100000.0
        return num
    except ValueError:
        return np.nan


# %%
# Cell-level comment: Execute dataframe-wide standardization logic across columns, roles, and tiers
# Strategy: Rename columns to snake_case, apply custom dictionary cleaning maps, and enforce precise clean dtypes.

# 1. Standardize column headers to snake_case
df.columns = df.columns.str.strip().str.lower()
df = df.rename(
    columns={
        "employee_id": "employee_id",
        "years_exp": "years_exp",
        "current_ctc": "current_ctc",
        "previous_ctc": "previous_ctc",
        "company_type": "company_type",
        "education_tier": "education_tier",
        "joining_year": "joining_year",
        "work_mode": "work_mode",
    }
)

# 2. Parse CTC columns using the custom scale-aware function
df["current_ctc_lpa"] = df["current_ctc"].apply(parse_ctc)
# Trap 2 Note: Keep previous_ctc as NaN for freshers to avoid skewing growth averages downstream
df["previous_ctc_lpa"] = df["previous_ctc"].apply(parse_ctc)

# 3. Clean and map messy Role titles to canonical structural variants
# Inspect unique values to map variants safely
role_map = {
    "SDE Backend": "SDE Backend",
    "Backend Engineer": "SDE Backend",
    "Backend Dev": "SDE Backend",
    "BE": "SDE Backend",
    "Backend Developer": "SDE Backend",
    "SDE Full-Stack": "SDE Full-Stack",
    "Full-Stack Engineer": "SDE Full-Stack",
    "Data Scientist": "Data Scientist",
    "Product Manager": "Product Manager",
    "DevOps Engineer": "DevOps Engineer",
    "Business Analyst": "Business Analyst",
    "UI/UX Designer": "UI/UX Designer",
    "Data Analyst": "Data Analyst",
}
df["role_clean"] = (
    df["role"].str.strip().map(role_map).fillna(df["role"].str.strip())
)

# 4. Standardize company type string casings
df["company_type_clean"] = df["company_type"].str.strip().str.capitalize()
# Re-map explicit anomalies if case-insensitivity missed them
df["company_type_clean"] = df["company_type_clean"].replace(
    {"Mnc": "MNC", "Mid-size": "Mid-size"}
)

# 5. Clean Education Tier labels to sharp categorical values
def clean_education(val):
    if pd.isna(val):
        return "Unknown"
    s = str(val).strip().upper().replace("-", "")
    if "1" in s or "T1" in s:
        return "Tier 1"
    if "2" in s or "T2" in s:
        return "Tier 2"
    if "3" in s or "T3" in s:
        return "Tier 3"
    return s


df["education_tier_clean"] = df["education_tier"].apply(clean_education)

# Drop explicit row duplicates, clean whitespaces from key string segments
df = df.drop_duplicates()
df["skills"] = df["skills"].fillna("").str.strip()

# Print verification values to guarantee clean processing state
print("--- VERIFYING CLEAN DATATYPES ---")
print(df.dtypes)
print("\n--- STANDARDIZED ROLE COUNTS ---")
print(df["role_clean"].value_counts())
```

Task 3

```python

role_dist = (
    df.groupby("role_clean")["current_ctc_lpa"]
    .agg(["median", "mean", "min", "max"])
    .sort_values(by="median", ascending=False)
)
print(role_dist)

# %% [markdown]
# > ### Insight (Q3.1)
# > Product Managers earn the highest median salary across the ecosystem, while Data Analysts sit at the lowest compensation bracket. Discrepancies between the mean and median within several technical tracks hint at highly paid outlier professionals shifting traditional baseline expectations upward.

# %% [markdown]
# ### Q3.2: Experience Curve for SDE Backend
# Bin engineering professionals into discrete cohorts to calculate growth rates.

# %%
# Comment: Calculate experiential progression bounds for Backend Engineers via custom bin cut structures
sde_backend_df = df[df["role_clean"] == "SDE Backend"].copy()

# Create experience bins safely
bins = [-1, 1, 3, 5, 100]
labels = ["0-1 years", "2-3 years", "4-5 years", "6+ years"]
sde_backend_df["exp_band"] = pd.cut(
    sde_backend_df["years_exp"], bins=bins, labels=labels
)

sde_exp_curve = (
    sde_backend_df.groupby("exp_band", observed=False)["current_ctc_lpa"]
    .median()
    .to_frame()
)
sde_exp_curve["pct_growth"] = sde_exp_curve["current_ctc_lpa"].pct_change() * 100
print(sde_exp_curve)

# %% [markdown]
# > ### Insight (Q3.2)
# > Compounded compensation growth rates accelerate significantly for SDE Backend engineers as they cross into the 4-5 and 6+ year experience brackets, proving that long-term domain specialization offers compounding career premiums.

# %% [markdown]
# ### Q3.3: Skill Premium for SDEs
# Isolating the explicit financial delta between SDEs who possess specialized skills and those who do not.

# %%
# Comment: Vectorized lookup to separate skill owners and evaluate corresponding salary shifts
sde_mask = df["role_clean"].str.contains("SDE", na=False)
sde_all_df = df[sde_mask].copy()

target_skills = ["AWS", "ML", "System Design", "Kubernetes"]
skill_premiums = {}

for skill in target_skills:
    has_skill = sde_all_df["skills"].str.contains(
        skill, case=False, na=False, regex=False
    )
    median_with = sde_all_df[has_skill]["current_ctc_lpa"].median()
    median_without = sde_all_df[~has_skill]["current_ctc_lpa"].median()
    premium_pct = (
        ((median_with - median_without) / median_without) * 100
        if median_without > 0
        else 0
    )

    skill_premiums[skill] = {
        "With": median_with,
        "Without": median_without,
        "Premium_%": premium_pct,
    }

skill_premium_df = pd.DataFrame(skill_premiums).T
print(skill_premium_df)

# %% [markdown]
# > ### Insight (Q3.3)
# > SDEs displaying expertise in System Design and Kubernetes command massive median salary bumps over their counterparts, making architecture and advanced orchestration highly lucrative focal points for personal upskilling.

# %% [markdown]
# ### Q3.4: Company-Type Premium
# Measuring market variations between standard Multinationals and domestic High-Growth Startups.

# %%
# Comment: Isolate market premiums across varying operational scales for SDE Backend developers
backend_comp_type = (
    sde_backend_df.groupby("company_type_clean")["current_ctc_lpa"]
    .median()
    .to_frame()
)

# Extract baseline numbers assuming typical standard values safely present
unicorn_median = backend_comp_type.loc["Unicorn", "current_ctc_lpa"]
mnc_median = backend_comp_type.loc["MNC", "current_ctc_lpa"]
premium_unicorn_vs_mnc = ((unicorn_median - mnc_median) / mnc_median) * 100

print(backend_comp_type)
print(f"\nUnicorn Premium over MNC: {premium_unicorn_vs_mnc:.2f}%")

# %% [markdown]
# > ### Insight (Q3.4)
# > High-growth Unicorns outpace traditional enterprise MNCs by paying a substantial premium for identical Backend Engineering talent, signaling where scaling pressures translate directly into aggressive cash compensation.

# %% [markdown]
# ### Q3.5: Underpaid Professionals Market Extraction
# Finding localized anomalies while protecting reliability via group size controls.

# %%
# Comment: Calculate localized cohort medians and find top underpaid outliers
# Safe creation of experience cohorts for full dataset to feed into group-level evaluation
df["exp_band"] = pd.cut(df["years_exp"], bins=bins, labels=labels)

# Groupby, transform, and filter securely out-of-the-box avoiding naive ChatGPT errors
group_cols = ["role_clean", "company_type_clean", "exp_band"]

df["group_median"] = df.groupby(group_cols, observed=False)[
    "current_ctc_lpa"
].transform("median")
df["group_count"] = df.groupby(group_cols, observed=False)[
    "current_ctc_lpa"
].transform("count")
df["gap_lpa"] = df["current_ctc_lpa"] - df["group_median"]

# Clean filters: extract only statistically significant cohorts containing 10 or more people
underpaid_candidates = df[df["group_count"] >= 10].sort_values(
    by="gap_lpa", ascending=True
)

print(
    underpaid_candidates[
        [
            "id",
            "role_clean",
            "company_type_clean",
            "years_exp",
            "current_ctc_lpa",
            "group_median",
            "gap_lpa",
        ]
    ].head(10)
)
```
